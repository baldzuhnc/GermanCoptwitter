#### Sentiment Analysis via IBM Watson API

```python
import os
import time
import json
import pandas as pd
from ibm_watson import NaturalLanguageUnderstandingV1
from ibm_watson.natural_language_understanding_v1 import Features, EntitiesOptions, KeywordsOptions, SentimentOptions

from ibm_cloud_sdk_core.authenticators import IAMAuthenticator
```

```python
# Authentication via IAM
authenticator = IAMAuthenticator('#YourAPI Key#')
service = NaturalLanguageUnderstandingV1(
    version='2018-03-16',
    authenticator=authenticator)

service.set_service_url('https://api.eu-de.natural-language-understanding.watson.cloud.ibm.com/instances/9f3785a9-5428-4299-aa49-9204ccd2639f')
```

```python
#Import data
feb = pd.read_csv("Tweetsfeb2.csv", header=0)

print(feb.head(n=5))
```

```python
# empty list for sentiment values
sentiment_values = []   
```

```python
#For loop API call


for i in range(len(feb)):
    
        # request is the current entry
        request = feb["full_text"][i]
        
        if len(request) > 10:
            # ibm watson sentiment analysis
            response = service.analyze(
                text=request,
                language='de',
                features=Features(sentiment=SentimentOptions(document=True))
        ).get_result()
        
        else:
            response = None # if minimum length not fulfilled

        # add value to list
        sentiment_values.append(response)
    
    
```

```python
sentiment_values
```

```python
# Results/Export
with open("Februar2sentidump.json", "w") as json_file:
    json.dump(sentiment_values, json_file)
```
